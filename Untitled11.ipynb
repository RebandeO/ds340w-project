```python
import pandas as pd

# Specify the file path
file_path = 'tweet data .xlsx'  # Change 'your_file.xlsx' to the path of your Excel file

# Load the Excel file into a pandas DataFrame
df = pd.read_excel(file_path)

# Print the first few rows of the DataFrame to verify the data is loaded correctly
print(df.head())
```

```python
# Convert the 'created_at' column to datetime format
df['created_at'] = pd.to_datetime(df['created_at'], format='%a %b %d %H:%M:%S +0000 %Y')

# Extract the date part from the datetime column
df['created_at'] = df['created_at'].dt.date

print(df)
```

```python
# prompt: Make the full_text column a string

df['full_text'] = df['full_text'].astype(str)
```

```python
# prompt: using this list and a team column based on the user/name column teams = {
#     "Lebron James": "Los Angeles Lakers",
#     "Joel Embiid": "Philadelphia 76ers",
#     "Kevin Durant": "Phoenix Suns",
#     "Stephan Curry":"Golden State Warriors",
#     "Giannis Antetokounmpo":"Milwaukee Bucks",
#     "Spencer Dinwiddle":"Los Angeles Lakers",
#     "Luka Doncic":"Dallas Mavericks",
#     "James Harden":"Los Angeles

teams = {
    "Lebron James": "Los Angeles Lakers",
    "Joel Embiid": "Philadelphia 76ers",
    "Kevin Durant": "Phoenix Suns",
    "Stephan Curry":"Golden State Warriors",
    "Giannis Antetokounmpo":"Milwaukee Bucks",
    "Spencer Dinwiddle":"Los Angeles Lakers",
    "Luka Doncic":"Dallas Mavericks",
    "James Harden":"Los Angeles Clippers"
}

df['team'] = df['user/name'].map(teams)

df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def analyze_sentiment(text):
    analyzer = SentimentIntensityAnalyzer()
    sentiment_scores = analyzer.polarity_scores(text)
    compound_score = sentiment_scores['compound']

    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['full_text'].apply(analyze_sentiment)

print(df)
```

```python
# Assuming merged_df is the DataFrame containing the merged data

# Export the DataFrame to a CSV file
df.to_csv('new_text_data.csv', index=False)  # Set index=False to exclude row numbers in the CSV file

```
